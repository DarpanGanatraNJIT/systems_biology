```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def enzymatic_reactions(t, y, k_sc, k_cs, k_cp):
    S, C, E, P = y
    dS = -k_sc*S*E + k_cs*C
    dC = k_sc*S*E - C*(k_cs + k_cp) 
    dE = -k_sc*S*E + (k_cs + k_cp)*C
    dP = k_cp*C
    return dS, dC, dE, dP

def output(k_sc, k_cs, k_cp, t_len):
    S0, C0, E0, P0 = 1, 0, 0.2, 0 
    y0 = S0, C0, E0, P0
    ret = solve_ivp(enzymatic_reactions, [0, t_len], y0, args = (k_sc, k_cs, k_cp))
    return ret

def plot_output(k_sc, k_cs, k_cp, t_len):
    ret_vals = output(k_sc, k_cs, k_cp, t_len)
    S, C, E, P = ret_vals.y
    plt.figure(figsize = (14,7))
    plt.plot(ret_vals.t, S, label = 'S')
    plt.plot(ret_vals.t, C, label = 'C')
    plt.plot(ret_vals.t, E, label = 'E')
    plt.plot(ret_vals.t, P, label = 'P')
    plt.legend()
    plt.ylabel("Concentration")
    plt.xlabel("Time")
    plt.show()
    
plot_output(100, 10, 2, 1)
    
```
